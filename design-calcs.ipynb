```python
from math import sin, cos, radians
import pandas as pd

CIRCLE_RADIUS = 12
CIRCLE_CENTER_X = 0
CIRCLE_CENTER_Y = 0


def circlePosIterator():
    yield 12
    yield from range(1, 12, 1)


def calcComponentCoordinates(deg, r=CIRCLE_RADIUS, cx=CIRCLE_CENTER_X, cy=CIRCLE_CENTER_Y):
    x = cx + r * sin(radians(deg))
    y = cy + r * cos(radians(deg))
    return x, y


def calcForCompleteCircle(startDegrees, incrementDegrees):
    posIter = circlePosIterator()
    degIter = range(startDegrees, 360, incrementDegrees)
    data = [[pos, deg] + list(calcComponentCoordinates(deg)) for pos, deg in zip(posIter, degIter)]
    return pd.DataFrame(data, columns=['Position', 'Degrees', 'X coordinate', 'Y coordinate'])
```

```python
# Coordinates for LEDs (every 30 degrees - starting at 0)
calcForCompleteCircle(0, 30)
```

```python
# Coordinates for resistors (every 30 degrees - starting at 15)
calcForCompleteCircle(15, 30)
```
